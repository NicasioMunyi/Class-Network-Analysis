**social Network analysis**

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
fa =nx.read_edgelist("facebook_combined.txt" , create_using =nx.Graph(), nodetype=int)
print(nx.info(fa))
```

```python
nx.draw(fa)
```

**Betweenes Centrality**

```python
pos = nx.spring_layout(fa)
betCent = nx.betweenness_centrality(fa, normalized=True, endpoints=True)
node_color = [20000.0 * fa.degree(v) for v in fa ]
node_size = [v * 10000 for v in betCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(fa, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('off')
sorted(betCent, key=betCent.get, reverse=True)[:5]
```

**Degree Of Centrality**

```python
pos = nx.spring_layout(fa)
degCent = nx.degree_centrality(fa)
node_color = [20000.0 * fa.degree(v) for v in fa ]
node_size = [v * 10000 for v in degCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(fa, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('off')
sorted(degCent, key=degCent.get, reverse=True)[:5]
```

**Closeness Centrality**

```python
pos = nx.spring_layout(fa)
cloCent = nx.closeness_centrality(fa, )
node_color = [20000.0 * fa.degree(v) for v in fa ]
node_size = [v * 10000 for v in cloCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(fa, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('off')
sorted(cloCent, key=cloCent.get, reverse=True)[:5]
```

**EigenVector Centrallity**

```python

eigCent = nx.eigenvector_centrality(fa, )
node_color = [20000.0 * fa.degree(v) for v in fa ]
node_size = [v * 10000 for v in eigCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(fa, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('off')
sorted(eigCent, key=eigCent.get, reverse=True)[:5]
```

**Computing All neighbours in the network**

```python
from networkx.classes.function import all_neighbors
n = [1,10,20,50,80,100,200,1000,]
for i in range(len(n)):
  all_neighbors = list(nx.classes.function.all_neighbors(fa,n[i]))
  print("Neigbours for the node", str(n[i]), "==", str(all_neighbors))
```
